# Preprocessing

```python
from matplotlib import pyplot
from pandas import read_csv
import numpy as np
filename = "diabetes_binary_5050split_health_indicators_BRFSS2015.csv"
data = read_csv(filename)
data['bmi_group'] = pd.cut(data['BMI'], (0, 16, 18.5, 25, 30, 35, 40, np.inf), labels=[1, 2, 3, 4, 5, 6, 7])
data.BMI = data['bmi_group']
data.drop('bmi_group', axis=1, inplace=True)
data.BMI = data.BMI.astype('float')

data.plot(kind='box', subplots=True, layout=(3,10), sharex=False, sharey=False, figsize=(20,5))
pyplot.show()
#Boxplot interpretation: 75 and 25 percent of the values lie between the upper and lower limits, respectively.
#the lines show the min and max values. In some cases there are points which are 1,5 times of the box lenth far away from the box.
#these are called outliners
```

```python
from sklearn.model_selection import train_test_split

y, X = data['Diabetes_binary'], data[data.columns[1:]]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=45)

numerical_col = ["BMI"]
categorical_col = ["HighBP", "HighChol", "CholCheck", "Smoker", "Stroke",
       "HeartDiseaseorAttack", "PhysActivity", "Fruits", "Veggies",
       "HvyAlcoholConsump", "AnyHealthcare", "NoDocbcCost", "GenHlth",
       "MentHlth", "PhysHlth", "DiffWalk", "Sex", "Age", "Education",
       "Income"]

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
ct = ColumnTransformer(transformers=[('num', MinMaxScaler(), numerical_col)], remainder='passthrough') #('cat', OneHotEncoder(),categorical_col)

# with scaled datas
X_transformed = ct.fit_transform(X)
X_train_transformed, X_test_transformed, y_train_transformed, y_test_transformed = train_test_split(X_transformed, y, random_state=45)
```

```python
# With deletion of the 6 socioeconomic features

from sklearn.model_selection import train_test_split

data_exclu = data.drop(['AnyHealthcare','NoDocbcCost', 'Sex', 'Age', 'Income', 'Education'], axis = 1)
y_exclu, X_exclu = data_exclu['Diabetes_binary'], data_exclu[data_exclu.columns[1:]]
X_train_exclu, X_test_exclu, y_train_exclu, y_test_exclu = train_test_split(X_exclu, y_exclu, random_state=45)

ct = ColumnTransformer(transformers=[('num', MinMaxScaler(), numerical_col)], remainder='passthrough') #('cat', OneHotEncoder(),categorical_col)

# with scaled datas
X_transformed_exclu = ct.fit_transform(X_exclu)
X_train_transformed_exclu, X_test_transformed_exclu, y_train_transformed_exclu, y_test_transformed_exclu = train_test_split(X_transformed_exclu, y_exclu, random_state=45)
```

## Recursive feature elimination

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
print('All features: {}'.format(lr.score(X_train, y_train)))

rfe = RFE(lr, n_features_to_select=11, step=1)
rfe.fit(X_train, y_train)

print("Num Features: {}".format(rfe.n_features_))
print("Selected Features: {}".format(rfe.support_))
print("Feature Ranking: {}".format(rfe.ranking_))

L = []
for i in range(len(rfe.ranking_)):
    if rfe.ranking_[i] == 1:
        L.append(i)
print(L)
```

Results with RFE with only raw data (ranked by importance) : 0 2 10 1 13 6 3 5 17 18 9
